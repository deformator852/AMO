```python
#5.)Дослідити реальний час виконання обох алгоритмів на Python і побудувати графіки залежностей T insert(n) та Т buble(n) для n = [5,10,50,100,500,1000,2000,3000,4000,5000,10000,20000,50000,100000],згенерувавши відповідні цілочисельні масиви.
#Сортування бульбашкою:
import time
arr = [12, 0, 66, 8, 9, 5, 67, 13]

def check_time(callbackfn, algorithm_name, arr):
    start = time.time()
    callbackfn(arr)
    end = time.time()
    result = end-start
    print(f"Час витрачений на сортування {algorithm_name}: {result} мілісекунд")

def bubble_sort(arr):
    swap = False
    while swap != True:
        swap = True
        for i in range(len(arr) - 1):
            if arr[i] > arr[i + 1]:
                tmp = arr[i]
                arr[i] = arr[i + 1]
                arr[i + 1] = tmp
                swap = False

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

check_time(bubble_sort, "сортування бульбашкою", arr.copy())
check_time(insertion_sort, "сортування вставкою", arr.copy())

    

```
