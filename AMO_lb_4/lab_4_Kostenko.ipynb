```python
#Cамостійна робота 1 

#Завдання 1:"Оцінити асимптотичну складність алгоритму лінійного пошуку у О-нотації."
#В-дь:
#У лінійному пошуку алгоритм просто перебирає всі елементи послідовно, починаючи з першого і закінчуючи останнім, доки не знайде шуканий елемент або не перебере всі елементи.
#Такий підхід гарантує, що найгірший випадок відбувається, коли елемент знаходиться в кінці списку, і вам потрібно перевірити всі елементи, що призводить до O(n) часу виконання.
#Тому, асимптотична складність алгоритму лінійного пошуку в О-нотації = O(n)
```

```python
#Cамоcтійна робота 2

#Завдання:1 "Оцінити асимптотичну складність алгоритму бінарного пошуку пошуку у О-нотації"
#В-дь:
#Кількість кроків, необхідних для пошуку, дорівнює log₂(n), де "n" - це розмір масиву. Тобто, на кожному кроці ми зменшуємо кількість елементів для пошуку вдвічі.
#Тому асимптотична складність алгоритму бінарного пошуку - O(log n).
import pandas as pd
import time
```

```python
#Завдання:2 "написати функцію, яка б могла повертати датафрейм з наступним складом полів: (“n”, “time”), де n – розмір масиву для функції `linear_searh();"
#В-дь:
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def measure_linear_search_time(arr_size):
    # Створюємо масив з випадковими значеннями
    arr = list(range(arr_size))
    target = arr_size + 1  # Шукаємо значення, якого немає у масиві

    # Вимірюємо час виконання linear_search()
    start_time = time.time()
    linear_search(arr, target)
    end_time = time.time()
    execution_time = end_time - start_time

    return arr_size, execution_time

def generate_dataframe(n_values):
    data = []
    for n in n_values:
        n, execution_time = measure_linear_search_time(n)
        data.append([n, execution_time])

    df = pd.DataFrame(data, columns=["n", "time"])
    return df

# Приклад використання:
n_values = [10, 100, 1000, 10000]  # Розміри масиву, для яких вимірюємо час
result_df = generate_dataframe(n_values)
print(result_df)
```

```python
#Завдання: 3 "написати функцію, яка б могла повертати датафрейм з наступним складом полів: (“n”, “time”), де n – розмір масиву для функції `bin_searh();"
#В-дь:
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1
    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

def measure_binary_search_time(arr_size):
    arr = list(range(arr_size))
    target = arr_size + 1  

    start_time = time.time()
    binary_search(arr, target)
    end_time = time.time()
    execution_time = end_time - start_time

    return arr_size, execution_time

def generate_dataframe(n_values):
    data = []
    for n in n_values:
        n, execution_time = measure_binary_search_time(n)
        data.append([n, execution_time])

    df = pd.DataFrame(data, columns=["n", "time"])
    return df

n_values = [10, 100, 1000, 10000]  
result_df = generate_dataframe(n_values)
print(result_df)
```
